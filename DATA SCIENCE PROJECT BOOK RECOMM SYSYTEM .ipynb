```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.groupby('Book-Title').count()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_ratings_df,on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False)
```
